```python
import os
import pandas as pd
import numpy as np
import datetime

PROJECT_ROOT = os.path.join(os.getcwd(), '..')
DATA_DIR = os.path.join(PROJECT_ROOT,'data')
MODEL_PATH = os.path.join(PROJECT_ROOT,'model')
```

```python
def get_now():
    now = datetime.datetime.now()
    return '{0:%Y-%m-%d %H:%M:%S}'.format(now)
```

```python
reader = pd.read_csv(os.path.join(DATA_DIR, 'train.csv.zip'), compression='zip', chunksize=100000, parse_dates=['click_time'])
for i, chunk in enumerate(reader):
    positive = chunk[chunk['is_attributed'] == 1]
    negative = chunk[chunk['is_attributed'] == 0]
    output_positive = os.path.join(DATA_DIR, 'train_positive.csv')
    output_negative = os.path.join(DATA_DIR, 'train_negative.csv')
    if i == 0:
        if os.path.isfile(output_positive):
            os.remove(output_positive)
            os.remove(output_negative)
        header = True
    else:
        header = False
    positive.to_csv(output_positive, index=False, header=header, mode='a')
    negative.to_csv(output_negative, index=False, header=header, mode='a')
    print('[{}]Finish:output Data'.format(get_now()))
print('[{}]All Process FInished'.format(get_now()))
```
